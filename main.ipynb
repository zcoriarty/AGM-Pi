```python
# import libraries
import math
import numpy as np
from decimal import *
import itertools
```

```python
# set precision
getcontext().prec = 10000

# define variables as Decimal objects (for precision control)
a_current = Decimal(1)
two = Decimal(2) 
g_current = 1/(two.sqrt())
t_current = Decimal(1/4)
```

```python
def geometric_mean(n, a_current, g_current, t_current):
	# recursive geometric definitions
	a_next = (a_current + g_current) / 2
	g = Decimal(a_current * g_current)
	g_next = g.sqrt()

	t_next = t_current - Decimal(math.pow(2, n) * math.pow(a_next - a_current, 2))

	pi = Decimal(math.pow(a_next, 2)) / t_next
	not_correct = check_pi(pi)
	print("Iteration: ", n)
	print("Full value: ", pi)
	print("Correct Digits: ", str(pi)[:len(str(pi))-not_correct])
	if n > 2:
		return
	return geometric_mean(n+1, a_next, g_next, t_next)
```

```python
# Compare the generated pi value to the true pi value
def check_pi(generated_pi):
	with open("pi") as infile:
		for line in infile: # line by line approach to prevent reading the whole file
			leftover = check_batch(generated_pi, line) 
			if leftover > 0:
				return leftover
			generated_pi = Decimal(str(generated_pi)[leftover:])



# return the number of digits in the computed value that are incorrect
def check_batch(generated_pi, true_pi):
	correct = sum(1 for _ in itertools.takewhile(lambda t:t[0]==t[1], zip(str(generated_pi), str(true_pi))))
	if correct > 0:
		return len(str(generated_pi)) - correct

	return -1
```
